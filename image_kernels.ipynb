```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
Gambar = cv2.imread ('/content/drive/MyDrive/Foto/Gambar 5.jpeg')
plt.imshow(Gambar),plt.grid(False)

plt.title('Orginal image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#blur
vertical_gd = np.array([[0.0625, 0.125, 0.0625], [0.125, 0.25,0.125], [0.0625, 0.125, 0.0625]])
kernel_3 = np.ones((3,3),np.float32)/1000

#apply
filter_v = cv2.filter2D(Gambar[:, :, 2], -1, vertical_gd)
filtered_3 = cv2.filter2D(filter_v,-1,kernel_3)

#show
plt.subplot(221)
plt.title('Original')
plt.imshow(Gambar)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.show()

plt.subplot(222)
plt.title('Blur')
plt.imshow(filter_v, cmap='gray')
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.show()

plt.subplot(223)
plt.title('Kernel')
plt.imshow(filtered_3, cmap='gray')
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.show()

```
